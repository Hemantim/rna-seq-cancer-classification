## Results

The proposed pipeline demonstrates excellent performance for multi-class cancer
classification using RNA-Seq gene expression data.

After preprocessing and cross-validated feature selection, Logistic Regression
achieved an accuracy of 99.38% and a macro F1-score of 0.995 on the held-out test
set. Random Forest achieved comparable performance, with an accuracy of 98.76%
and a macro F1-score of 0.989.

Both models show near-perfect classification across all cancer types. The few
misclassifications observed occur primarily for LUAD samples, indicating partial
biological overlap with other cancer types. Overall, the results confirm that
RNA-Seq expression profiles contain highly discriminative information for cancer
tye prediction.


## Discussion

This study demonstrates that effective preprocessing and feature selection
enable highly accurate cancer classification using relatively simple models.
Despite its linear nature, Logistic Regression slightly outperformed Random Forest,
suggesting that the feature-selected data is largely linearly separable.

Optimizing feature selection using macro F1-score proved beneficial, as it ensured
balanced performance across all cancer types rather than favoring dominant classes.
The increase in the optimal number of selected genes reflects the need to retain
additional features to improve the classification of more challenging cancer types.

Feature importance analysis revealed that both models rely on a compact subset of
genes, with partial overlap between linear and non-linear importance rankings.
Clustering analysis further supported the supervised results, showing that samples
naturally group by cancer type, while also highlighting the intrinsic overlap between
certain cancers such as LUAD.
